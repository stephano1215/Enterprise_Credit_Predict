### 여기서부터 추가하시면 됩니다

```python
# 기업신용등급 가져오기
coName = []
coRating = []
ratingYear = []
ratingCo = []
startdate = '2015-01-01' #yyyy-mm-dd
enddate = '2018-12-31'#yyyy-mm-dd
for i in range(1, 10):
    cnt = 1
    page = str(i)
    url = 'http://www.fss.or.kr/fss/kr/bsn/cad/cad_rpt_l.jsp?p_ass_gbn=%C1%A4%B1%E2%C6%F2%B0%A1&p_target_type_kor=%B1%E2%BE%F7%BD%C5%BF%EB%C6%F2%B0%A1&p_end_date=' + enddate +'&p_target_co_nm=&p_condition_date=GONGSI_DT&p_company=&y=10&p_start_date=' + startdate +'&x=25&page=' + page
    soup = BeautifulSoup(requests.get(url).text, "html.parser")
    numOfPage = soup.find('span', {'class' : 'num'})
    table = soup.find('table', {'class' : 'bbs_list'})
    listt = table.find_all("tr")
    listt2 = table.find_all('td')
    for j in range(1,len(listt)):
        x = 1
        ratingCo.append(listt2[x].text)
        coName.append(listt[cnt].find("td", {"class":"tit"}).text)
        coRating.append(listt[cnt].find_all(text=re.compile("^[A-C]")))
        ratingYear.append(listt[cnt].find_all(text=re.compile("^[2]\d{3}")))
        cnt += 1
        x += 10
```

```python
df_ratingCo = pd.DataFrame(ratingCo, columns = ['RatingCo'])
df_coName = pd.DataFrame(coName, columns=['CompanyName'])
df_coRating = pd.DataFrame(coRating, columns = ['Rating'])
df_ratingYear = pd.DataFrame(ratingYear, columns = ['Day'])
```

```python
#데이터프레임 합치기
df = pd.concat([df_ratingCo, df_coName, df_coRating, df_ratingYear], axis=1)
df
```

```python
#중복 제거
df2 = df.drop_duplicates(['RatingCo', 'CompanyName', 'Day'], keep='first')
df2
```
