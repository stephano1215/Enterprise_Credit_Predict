## 정규화 후 랜덤포레스트

```python
#결정 트리의 주요 단점은 훈련 데이터에 과대적합되는 경향이 있다는 것이었다.
#랜덤 포레스트는 이 문제를 회피할 수 있는 방법

#엑스트라 트리는 랜덤 포레스트 모델의 변종
#엑스트라 트리는 포레스트 트리의 각 후보 특성을 무작위로 분할하는 식으로 무작위성을 증가
#결정트리를 사용하는 랜덤포레스트와 차이를 보임
#랜덤포레스트와 마찬가지로 다중공선성 문제해결, 과대적합 악화를 해줌.


print(__doc__)

import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.ensemble import ExtraTreesClassifier


# 포레스트 만들고 변수 중요도 계산, 노드 개수 500
forest = ExtraTreesClassifier(n_estimators=500,
                              random_state=0)

forest.fit(Xd, y)
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1] #오름차순정렬을 내림차순으로

# 변수중요도 보여줌
print("Feature ranking:")

# 상위 3개만 보여줌
for f in range(Xd.shape[1]):
    if f < 3:
        print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))

# 변수중요도 시각화
plt.figure(figsize=(200,100))
plt.title("Feature importances",fontsize = 500,color ='y')
plt.bar(range(Xd.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(Xd.shape[1]), indices)
plt.xticks(fontsize = 120, color = 'black')
plt.yticks(fontsize = 130, color = 'black')
plt.xlim([-1, Xd.shape[1]])
plt.show()
```
