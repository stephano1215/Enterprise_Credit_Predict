### 재무정보 파싱

```python
from pyfinmod.yahoo_finance import YahooFinanceParser
import numpy as np
from pandas import Series, DataFrame


def income(x):
    parser = YahooFinanceParser(x)
    df=parser.get_dataframe('income-statement')
    df2=np.transpose(df)
    df2.insert(0,"companyname",x)
    return(df2)

def balance(x):
    parser = YahooFinanceParser(x)
    df=parser.get_dataframe('balance-sheet')
    df2=np.transpose(df)
    #df2.insert(0,"companyname",x)
    return(df2)
```

### 수익성

```python
import sys
# 매출총이익 / 유동자산 * 100
a = df_csv["Gross Profit"] #매출총이익 (Gross Fit)
b = df_csv["Total Current Assets"] #유동자산
result = a/b *100
df = pd.DataFrame(result)
df2 = df.rename(columns = {0: "Gross Profit/Total Current Assets"})
df_final = pd.concat([df_csv, df2], axis = 1)
df_final.head()
```

```python
#총자산수익율 = 당기순이익 / 자산 * 100
a = df_csv["Net Income"] #당기순이익
b = df_csv["Total Assets"] #자산
result = a/b *100
df = pd.DataFrame(result)
df2 = df.rename(columns = {0: "Net Income/Total Assets"})
df_final = pd.concat([df_final, df2], axis = 1)
df_final
```

```python
#자본금순이익율 = 당기순이익 / 자본금 * 100
a = df_csv["Net Income"] #당기순이익
b = df_csv["Total Assets"] - df_csv["Total Liabilities"] #자본(자산 - 부채)
result = a/b *100
df = pd.DataFrame(result)
df2 = df.rename(columns = {0: "Net Income/Total Capital"})
df_final = pd.concat([df_final, df2], axis = 1)
df_final
```

```python
#영업이익율 = 영업이익 / 매출액 * 100
a = df_csv["Gross Profit"]
b = df_csv["Total Revenue"]
result = a/b * 100
df = pd.DataFrame(result)
df2 = df.rename(columns = {0: "Gross Profit/Total Revenue"})
df_final = pd.concat([df_final, df2], axis = 1)
df_final
```

```python
#총자산순이익율 = 당기순이익 / 총자산 * 100
a = df_csv["Net Income"] #당기순이익
b = df_csv["Total Assets"] #자산
result = a/b * 100
df = pd.DataFrame(result)
df2 = df.rename(columns = {0: "Net Income/Total Assets"})
df_final = pd.concat([df_final, df2], axis = 1)
df_final
```

### 안정성

```python
#(부채총계 - 이익잉여금) / 유형자산 * 100
a = df_csv["Total Liabilities"] - df_csv["Retained Earnings"]
b = df_csv["Net Tangible Assets"]
result = a/b * 100
df = pd.DataFrame(result)
df2 = df.rename(columns = {0: "TL - RE/Net Tangible Assets"})
df_final = pd.concat([df_final, df2], axis = 1)
df_final
```

```python
#매출총이익 / 이자비용 : 매출총이익 / 이자비용 * 100
#마이너스 빼야 할듯!
a = df_csv["Gross Profit"]
b = df_csv["Interest Expense"] * -1
result = a/b * 100
df = pd.DataFrame(result)
df2 = df.rename(columns = {0: "Gross Profit/Interest Expense"})
df_final = pd.concat([df_final, df2], axis = 1)
df_final
```

```python
#고정부채비율 : 고정부채 / 자본 * 100
a = df_csv["Long Term Debt"] #고정부채(비유동부채)
b = df_csv["Total Assets"] - df_csv["Total Liabilities"]# 자본총계
result = a/b * 100
df = pd.DataFrame(result)
df2 = df.rename(columns = {0: "Long Term Debt/Total Assets"})
df_final = pd.concat([df_final, df2], axis = 1)
df_final
```

```python
#이익잉여금 / 부채총계 : 이익잉여금 / 부채 * 100
a = df_csv["Net Income"] # 이익잉여금
b = df_csv["Total Liabilities"] # 부채총계
result = a/b * 100
df = pd.DataFrame(result)
df2 = df.rename(columns = {0: "Net Income/Total liabilities"})
df_final = pd.concat([df_final, df2], axis = 1)
df_final
```

```python
#이익잉여금 / 자산총계 : 이익잉여금 / 자산 * 100
a = df_csv["Net Income"]  # 이익잉여금
b = df_csv["Total Assets"]# 자산총계
result = a/b * 100
df = pd.DataFrame(result)
df2 = df.rename(columns = {0: "Net Income/Total Assets"})
df_final = pd.concat([df_final, df2], axis = 1)
df_final
```

### 활동성

```python
#총자산회전율 = 매출액 / 자산총계
a = df_csv["Total Revenue"] #매출액
b = df_csv["Total Assets"]# 자산총계
result = a/b * 100
df = pd.DataFrame(result)
df2 = df.rename(columns = {0: "Total Revenue/Total Assets"})
df_final = pd.concat([df_final, df2], axis = 1)
df_final
```

```python
#총자본회전율 = 매출액 / 자본총계
a = df_csv["Total Revenue"] #매출액
b = df_csv["Total Assets"] - df_csv["Total Liabilities"]# 자본총계
result = a/b * 100
df = pd.DataFrame(result)
df2 = df.rename(columns = {0: "Total Revenue/Total Capital"})
df_final = pd.concat([df_final, df2], axis = 1)
df_final
```

```python
#순운전자본회전율 = 매출액 / 유동자산 - 유동부채
a = df_csv["Total Revenue"] #매출액
b = df_csv["Total Current Assets"] #유동자산
c = df_csv["Total Current Liabilities"] # 유동부채
result = a / (b-c)
df = pd.DataFrame(result)
df2 = df.rename(columns = {0: "Total Revenue/TCA - TCL"})
df_final = pd.concat([df_final, df2], axis = 1)
df_final
```

### 유동성

```python
#당좌비율 = 당좌자산 / 유동부채 * 100
a = df_csv["Total Current Assets"]
b = df_csv["Total Current Liabilities"] # 유동부채
result = a/b * 100
df = pd.DataFrame(result)
df2 = df.rename(columns = {0: "TCA/TCL"})
df_final = pd.concat([df_final, df2], axis = 1)
df_final
```

```python
#현금비율 = (현금및현금등가물 + 단기금융상품) / 유동부채
a = df_csv["Cash And Cash Equivalents"]
b = df_csv["Short Term Investments"]
c = df_csv["Total Current Liabilities"] # 유동부채
result = (a+b) / c * 100
df = pd.DataFrame(result)
df2 = df.rename(columns = {0: "Cash + STI/TCL"})
df_final = pd.concat([df_final, df2], axis = 1)
df_final
```

## 5. Weighted average cost of capital and DCF

```python
from pyfinmod.yahoo_finance import YahooFinanceParser
from pyfinmod.wacc import wacc
from pyfinmod.ev import dcf

parser = YahooFinanceParser('000660.KS')
e = parser.get_value('Market Cap')
beta = parser.get_value('Beta (3Y Monthly)')
income_statement = parser.get_dataframe('income-statement')
balance_sheet = parser.get_dataframe('balance-sheet')

WACC= wacc(e, balance_sheet, income_statement, beta,
                 risk_free_interest_rate=0.02,
                 market_return=0.08)
DCF=dcf(FCF, WACC, short_term_growth=0.08, long_term_growth=0.04)

print(WACC)
print()
print()
print(DCF)
print()
print()
print(FCF)
```
